## Analysing nuclear accidents

```python
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

This dataset compiles all the nuclear accidents worldwide until 2016 (source: https://innovwiki.ethz.ch/index.php/Nuclear_events_database).

```python
df = pd.read_csv("NuclearPowerAccidents2016.csv")
df.head()
```

```python
df.columns
```

## 1) Which region has suffered more nuclear accidents?
**(Note that the 'Region' column is empty for North America)**

```python
df['Region'].value_counts()
```

```python
df[df['Region'] == 'WE']
```

## 2) Which are the fatalities per accident per region?

```python
df.groupby('Region')['Fatalities'].value_counts()
```

## 3) Which accidents took place in Western Europe

```python
western_europe = df[df['Region'] == 'WE']
western_europe
```

## 4) Which is the accident with the higher victims toll in WE?

```python
western_europe.sort_values(by='Fatalities', ascending=False)
```

## 5) What about the most expensive (in WE)?

```python
western_europe.sort_values(by='Cost (millions 2013US$)', ascending=False)
```

## 6) Which location in WE concentrates the highest number of accidents?

```python
western_europe.groupby('Location')['Fatalities'].value_counts().sort_values(ascending=False)
```

## 7) Replacing NaN with North America

```python
north_america = df[df['Region'].isnull()]
north_america
```

```python
north_america.info()
```

```python
north_america1 = north_america.fillna("NA")
north_america1
```

```python
north_america['Region'] = north_america['Region'].fillna("NA")
north_america
```

There are two accidents (one in Civaux, France, and another in Sellafiedl, UK), where the Region column has a NaN value. Since this has been now replaced by NA, those two rows need to be removed from the `north_america` subdataset to prevent false results.

```python
north_america_2 = north_america[(north_america['Location'] != 'Civaux, France') & (north_america['Location'] != 'Sellafield, UK')]
north_america_2
```

## 8) How many accidents in NA?

```python
north_america_2['Location'].value_counts().sum()
```

## 9) Accidents' severity (INES and Smyth magnitude)

This DataFrame includes two different ways of quantifying the severity of a nuclear accident: INES (International Nuclear Event Scale, developed by the International Atomic Energy Agency, http://www-ns.iaea.org/tech-areas/emergency/ines.asp) and Smyth Magnitude (created by David Smythe, it wants to be a more comprehensive measure of the impact of nuclear accidents regarding the radioactivity they release to the atmosphere, http://www.davidsmythe.org/nuclear/accidents.htm). In the first scale, 7 is the highest level. The second is a logarithmic scale. 

```python
df.sort_values(by="INES", ascending=False).head(10)
```

```python
df.sort_values(by="Smyth Magnitude", ascending=False).head(10)
```

## 10) Which region concentrates the highest number of high Smythe Magnitude accidents?

We define the "high Smythe Magnitude accidents" as those over 4 in that scale.

```python
smyth_magnitude_over_4 = df[df['Smyth Magnitude'] > 4]
smyth_magnitude_over_4.groupby('Smyth Magnitude')['Region'].value_counts().sort_values(ascending=False)
```

Four out of seven high severity accidents are located in Eastern Europe.

## 11) When did the most severe accidents happen?

```python
smyth_magnitude_over_4 = df[df['Smyth Magnitude'] > 4]
smyth_magnitude_over_4.groupby('Smyth Magnitude')['Date'].value_counts().sort_values(ascending=False)
```

## 12) Did the most severe accidents have the highest victim toll?

```python
smyth_magnitude_over_4 = df[df['Smyth Magnitude'] > 4]
smyth_magnitude_over_4.groupby('Smyth Magnitude')['Fatalities'].value_counts().sort_values(ascending=False)
```

Not necessarily, as the figures show.

## Graph 1: Is there a correlation between how much an accident costs and how many people it affects?

```python
df.plot(kind='scatter', x='Cost (millions 2013US$)', y='Fatalities', alpha=0.2)
```

## Graph 2: is there any correlation between the date of the accident and its severity?

```python
smyth_magnitude_over_4.sort_values(by='Smyth Magnitude', ascending=False).plot(kind='barh', x='Date', y='Smyth Magnitude', legend=False)
```
